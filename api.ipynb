```python
import requests
import pandas as pd
import os

# Base URL for the API

```

```python
BASE_URL = "https://api.openf1.org/v1"
```

```python
def get_latest_session_key():
    response = requests.get(f"{BASE_URL}/sessions?meeting_key=latest")
    response.raise_for_status()
    return response.json()[0]['session_key']
```

```python
def fetch_data(endpoint, session_key):
    url = f"{BASE_URL}/{endpoint}?session_key={session_key}"
    response = requests.get(url)
    response.raise_for_status()
    return pd.DataFrame(response.json())
```

```python
def save_data(df, name):
    filepath = f"f1_data/{name}.csv"
    df.to_csv(filepath, index=False)
    print(f"Saved {name} data to {filepath}")
```

```python
session_key = get_latest_session_key()
laps_df = fetch_data("laps", session_key)
save_data(laps_df, "laps")
```
